# Web Traffic Time Series Forecasting(EDA)

1. 대회 개요
- 약 145,000개의 Wikipedia 문서별 하루 조회 수(time series) 데이터를 사용해, 향후 일정 기간의 트래픽을 예측하는 문제
- 두 단계(Stage 1: ~2016년말, Stage 2: 데이터 확장)로 구성된 대회 구조\




2. 데이터 구성
- train 파일: 각 페이지별 날짜별 조회수가 열로 제공됨 (예: Page 컬럼 + 날짜 컬럼들).
- key 파일: 각각의 페이지 코드나 메타데이터 매핑용 정보 포함 \




3. 평가 지표
- SMAPE (Symmetric Mean Absolute Percentage Error)로 계산되며, 예측과 실제 간의 대칭적 오차 비율을 측정\




4. 주요 접근 방식
- 고전 시계열 모델: AR, MA, ARIMA 등을 사용한 베이스라인 분석 
- 머신러닝/딥러닝 혼합 모델
    - ARIMA + LSTM 앙상블 방식 
    - LSTM, CNN, seq2seq 기반의 트랜스포머 아키텍처 활용한 고급 모델 .

### Access type
- all-access : 모든 경로(웹 + 모바일웹 + 모바일앱) 포함
- desktop : 데스크탑 브라우저에서 접근한 경우
- mobile-web : 모바일 브라우저(앱이 아닌 웹)에서 접근한 경우

### Agent
- all-agents : 모든 에이전트 포함
- user : 실제 사용자 (사람)
- spider : 크롤러, 봇, 검색엔진 로봇 등 자동화된 시스템
- automated : 명확하게 자동화된 비사용자 봇 (spider보다 더 제한적 정의)

## Splitting data based on different access types and different agents.

```python
# Finding number of access types and agents
access_types = []
agents = []

for column in df.columns:
    access_type = column.split("_")[-2]
    agent = column.split("_")[-1]
    access_types.append(access_type)
    agents.append(agent)
```

```python
# access types 갯수 세기
from collections import Counter

access_dict = Counter(access_types)
access_dict
```

```python
print("Number of topics with all-access type:", access_dict["all-access"])
print("Number of topics with desktop access:", access_dict["desktop"])
print("Number of topics with mobile-web access:", access_dict["mobile-web"])
```

```python
access_df = pd.DataFrame({"Access type" : access_dict.keys(),
                         "Number of columns" : access_dict.values()})

access_df
```

```python
# Counting agents
agents_dict = Counter(agents)
agents_dict
```

```python
print("Number of topics with spider as agent:", agents_dict["spider"])
print("Number of topics with all-agents as agent:", agents_dict["all-agents"])
```

```python
agents_df = pd.DataFrame({"Agent" : agents_dict.keys(),
                          "Number of columns" : agents_dict.values()})

agents_df
```

```python
# access type의 결측치 있는 컬럼 확인
def count_null_columns(pattern):
    pattern_columns = [column for column in df.columns if pattern in column]
    return len(df[pattern_columns].isnull().sum()[df[pattern_columns].isnull().sum() > 0])

no_of_cols_with_nulls = [count_null_columns(access_type) for access_type in access_df["Access type"]]

access_df["No of columns with nulls"] = no_of_cols_with_nulls
access_df
```

```python
# agent의 결측치 있는 컬럼 확인
def count_null_columns(pattern):
    pattern_columns = [column for column in df.columns if pattern in column]
    return len(df[pattern_columns].isnull().sum()[df[pattern_columns].isnull().sum() > 0])

no_of_cols_with_nulls = [count_null_columns(agent) for agent in agents_df["Agent"]]

agents_df["No of columns with nulls"] = no_of_cols_with_nulls
agents_df
```

```python
# access type의 null value의 비율
access_df["% of nulls"] = access_df["No of columns with nulls"] / access_df["Number of columns"] * 100
access_df
```

Percentage of missing values in each access type is almost same.\

## 여기부터 내 맘대로 분석 시작

```python
df_new = pd.read_csv("train_1.csv")
df_log = df_new.copy()

df_log.iloc[:, 1:] = np.log1p(df_log.iloc[:, 1:]) # page열 제외

# np.log1p(x) = log(1 + x) -> 로그 변환을 안정적으로 해줌
# np.log는 입력값이 0일때 계산이 불가능
```

```python
# 원본과 로그 변환 비교
page_data = df.iloc[0, 1:].astype(float)
page_log = df_log.iloc[0, 1:]
page_log
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)
plt.plot(page_data.values)
plt.title("Original View Count")

plt.subplot(1, 2, 2)
plt.plot(page_log.values)
plt.title("Log1p Transformed")
plt.show()
```

```python
df_log_interp = df_log.copy()
df_log_interp.iloc[:, 1:] = df_log_interp.iloc[:, 1:].interpolate(method='linear', axis=1, limit_direction='both')
```

interpolate => 결측치를 보간법을 이용해 메꿈\
axis = 1이면 col 방향으로 메꿈(시간순)\

보간법 적용 후 결측치가 652개 -> 각 날짜별로 652개 페이지가 조회수가 없음

```python
# 전체 날짜에 대해 거의 다 NaN인 페이지(행) 제거

nan_threshold = df_log.shape[1] * 0.9  # 예: 90% 이상이 NaN이면 제거
df_clean = df_log[df_log.isnull().sum(axis=1) < nan_threshold]
```

## 시계열 예측 시작

### ARIMA
AutoRegressive INtergrated Moving Average\
선형 회귀 기반 + 시차 항

- 장점\
수학적으로 해석 가능\
통계 기반 정밀 모델링

- 단점
복잡한 튜닝 필요\
비정상 데이터 처리 필요





### Prophet
facebook Prophet\
추세 + 계절성 + 휴일 효과 조합

- 장점\
자동화 용이\
계절성/추세/이상치 robust\
실무에 빠르게 적용

- 단점\
복잡한 상호작용 반영 어려움\
일부 패턴 단순화

```python
# 최근 7일의 평균
rolling_pred = df_clean.iloc[:, -7:].mean(axis=1)
rolling_pred
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

series = df_clean.iloc[0, 1:].dropna()  # 한 페이지 선택
model = ARIMA(series, order=(1,1,1))    # 예시 파라미터
model_fit = model.fit()
forecast = model_fit.forecast(steps=30)

plt.plot(series[-100:], label='history')
plt.plot(range(len(series), len(series)+30), forecast, label='forecast')
plt.legend()
plt.show()
```

ARIMA 모델을 사용해서 1개 페이지의 향수 30일 조회수를 예측하는 코드

로그변환 한거 때문에 forecast 그래프가 이상함

```python
# 예측값 복원
forecast_real = np.expm1(forecast)

# 시계열도 복원
series_log = df_clean.iloc[0, 1:].astype(float).dropna()
series_real = np.expm1(series_log)

# 시각화
plt.figure(figsize=(12,5))
plt.plot(series_real[-100:], label='Actual')
plt.plot(range(len(series_real), len(series_real)+30), forecast_real, label='Forecast')
plt.legend()
plt.title("ARIMA 예측 결과 (로그 복원)")
plt.ylim(0, np.percentile(series_real, 95))  # y축 조정 (optional)
plt.show()
```
